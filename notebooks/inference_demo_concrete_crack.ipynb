```python
from mmseg.apis import init_segmentor, inference_segmentor, show_result_pyplot, inference_segmentor_sw
from mmseg.core.evaluation import get_palette
```

```python
config_file = '/home/user/UOS-SSaS Dropbox/05. Data/03. Checkpoints/021. Korean_Express_Corporation/2022.08.18 cgnet_2048x2048_60k_cityscapes_class_uniform_sample_crack.py/cgnet_2048x2048_60k_cityscapes_class_uniform_sample_crack.py'
checkpoint_file = '/home/user/UOS-SSaS Dropbox/05. Data/03. Checkpoints/021. Korean_Express_Corporation/2022.08.18 cgnet_2048x2048_60k_cityscapes_class_uniform_sample_crack.py/iter_60000.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_segmentor(config_file, checkpoint_file, device='cuda:0')
```

```python
# test a single image
img = '/home/user/UOS-SSaS Dropbox/05. Data/02. Training&Test/021. Korean_Express_Corporation/Fine Labeling (2021-2022)/road_crack/leftImg8bit/val/P-005-1-자유로ICR-A교_s000192000__3072,1024)__leftImg8bit.png'
result = inference_segmentor(model, img)

# show the results
show_result_pyplot(model, img, result, [[0, 0, 0], [255, 0, 0]])
```

```python
from glob import glob 
import os 
import cv2

img_folder = '/home/user/UOS-SSaS Dropbox/05. Data/02. Training&Test/021. Korean_Express_Corporation/Fine Labeling (2021-2022)/HaWonGok_1Gyo/1차로'

img_list = glob(os.path.join(img_folder, '*.jpg'))

for img_path in img_list:
    result = inference_segmentor_sw(model, img_path, window_size=2048)
    org_img = cv2.imread(img_path)
    color_img = org_img
    color_img[result[0] == 1, :] = [0, 0, 255]
    save_path = img_path.replace('.jpg', '_result.png')
    cv2.imwrite(save_path, color_img)

    print(img_path)
```
